```python
from lib.ekyn import *
import pandas as pd
```

```python
ids = get_ekyn_ids()
```

```python
for i,idx in enumerate(ids[:8]):
    for j,condition in enumerate(['Vehicle','PF']):
        X,y = load_ekyn_pt(idx=idx,condition=condition)
        print(X.shape)
        df = pd.DataFrame(X.numpy())
        df[5000] = y.argmax(axis=1).numpy()
        df.to_csv(f'bdhsc_2024/stage1_labeled/{i}_{j}.csv',index=False)
```

```python
for i,idx in enumerate(ids[8:]):
    for j,condition in enumerate(['Vehicle','PF']):
        X,y = load_ekyn_pt(idx=idx,condition=condition)
        print(X.shape)
        df = pd.DataFrame(X.numpy())
        df.to_csv(f'bdhsc_2024/stage2_unlabeled/{i+8}_{j}.csv',index=False)
```
