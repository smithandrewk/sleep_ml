```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv(f'data.csv')
df = df.drop('Unnamed: 0',axis=1)
X = df.to_numpy()
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
for i,element in enumerate(range(10,20)):
    print(i,element)
```

```python
def kmeans_clustering(number_of_clusters):
    number_of_clusters = 3
    center_1 = np.random.randn(2)
    center_2 = np.random.randn(2)
    center_3 = np.random.randn(2)
    plt.scatter(X[:,0],X[:,1])
    plt.scatter(*center_1)
    plt.scatter(*center_2)
    plt.scatter(*center_3)
    last_center_1 = center_1
    last_center_2 = center_2
    last_center_3 = center_3
    for i in range(20):
        cluster_1 = []
        cluster_2 = []
        cluster_3 = []

        for point in X:
            dist_1 = ((center_1[0]-point[0])**2 + (center_1[1]-point[1])**2)**.5
            dist_2 = ((center_2[0]-point[0])**2 + (center_2[1]-point[1])**2)**.5
            dist_3 = ((center_3[0]-point[0])**2 + (center_3[1]-point[1])**2)**.5
            distances = [dist_1,dist_2,dist_3]
            shortest_idx = 0
            for i,distance in enumerate(distances):
                if distance < distances[shortest_idx]:
                    shortest_idx = i
            if shortest_idx == 0:
                cluster_1.append(point)
            elif shortest_idx == 1:
                cluster_2.append(point)
            else:
                cluster_3.append(point)

        center_1 = np.array(cluster_1).mean(axis=0)
        center_2 = np.array(cluster_2).mean(axis=0)
        center_3 = np.array(cluster_3).mean(axis=0)
        if (center_1 == last_center_1).all() and (center_2 == last_center_2).all() and (center_3 == last_center_3).all():
            print("done")
            break
        last_center_1 = center_1
        last_center_2 = center_2
        last_center_3 = center_3
        print(center_1,center_2,center_3)
        plt.figure()
        plt.scatter(np.array(cluster_1)[:,0],np.array(cluster_1)[:,1])
        plt.scatter(np.array(cluster_2)[:,0],np.array(cluster_2)[:,1])
        plt.scatter(np.array(cluster_3)[:,0],np.array(cluster_3)[:,1])
        plt.scatter(*center_1)
        plt.scatter(*center_2)
        plt.scatter(*center_3)
```

```python
kmeans_clustering(3)
```
