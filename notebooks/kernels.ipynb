```python
from lib.models import ResNet
import torch
import matplotlib.pyplot as plt

model = ResNet(5000)
model.load_state_dict(torch.load('alpha/model.pt'))
```

```python
import matplotlib.pyplot as plt
weights = model.block1.c1.weight
fig,axes = plt.subplots(nrows=len(weights),ncols=1,figsize=(32,18))
for i in range(len(weights)):
    axes[i].plot(weights[i].squeeze().detach().cpu())
plt.savefig('kernels.jpg',dpi=500)
```
