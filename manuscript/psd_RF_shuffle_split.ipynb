```python
from lib.ekyn import *
from lib.utils import *

from sklearn.model_selection import train_test_split
import joblib
from sklearn.ensemble import RandomForestClassifier
```

```python
X,y = load_psd_label_pairs_windowed(get_ekyn_ids())
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.2,shuffle=True,random_state=0)
```

```python
clf = RandomForestClassifier(random_state=0)
```

```python
clf.fit(X_train,y_train)
```

```python
# joblib.dump(clf,'psd_RF_shuffle_split.joblib')
```

```python
# saved state
clf = joblib.load('psd_RF_shuffle_split.joblib')
```

```python
y_pred = clf.predict(X_test)
metrics(y_pred=y_pred,y_true=y_test)
```
