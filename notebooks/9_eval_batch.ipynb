```python
import os
import matplotlib.pyplot as plt
import json
import pandas as pd
```

```python
df = pd.DataFrame()
for project in os.listdir(f'../projects'):
    with open(f'../projects/{project}/config.json','r') as f:
        CONFIG = json.load(f)
    df[project] = CONFIG
df = df.T
df.sort_values(by='BEST_DEV_F1',ascending=False)
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(16,5),sharey='row',sharex=True)
for project in os.listdir(f'../projects'):
    if project.split('_')[2] not in ['b2']:
        continue
    print(project)
    with open(f'../projects/{project}/config.json','r') as f:
        CONFIG = json.load(f)
        ax[0,0].plot(CONFIG['TRAINLOSSI'],label=project)
        ax[0,1].plot(CONFIG['DEVLOSSI'],label=project)
        ax[1,0].plot(CONFIG['TRAINF1'],label=project)
        ax[1,1].plot(CONFIG['DEVF1'],label=project)
ax[0,0].set_title('train')
ax[0,1].set_title('dev')
ax[0,0].set_ylabel('loss')
ax[1,0].set_ylabel('f1')
ax[1,0].set_xlabel('epoch')
ax[1,1].set_xlabel('epoch')
ax[1,0].set_ylim([.7,1])
ax[0,0].set_ylim([.5,1.5])
plt.legend()
```
